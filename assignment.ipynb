<h1>3.Logical operators</h1>

Logical operators combine multiple condition

There are total three logical operators:-

1.AND:-All conditions must be true---->return true

2.OR:-At least one condition must be true---->return true otherwise false

3.NOT:-Reverse the boolean value---->true becomes false and false becomes true



```python
a = 10
b = 20

print(a < b and b > 15)
print(a > b and b > 15)
```

```python
x=-3
y=-1
print(x>y)
```

```python
a = 10
b = 20

print(a > b or b > 15)
print(a > b or b < 15)
```

```python
a = 10.5
b = -20

print(a > b or b > 15)
print(a > b or b < 15)
```

```python
a = 10

print(not(a > 5))
print(not(a < 5))
```

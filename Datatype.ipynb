List:-

List is the ordered,mutable collection of element in python

List is the changeable we can modify list

List is represented by [ ]

```python
l=[10,20,30]
print(l)
```

Tuple:-

Tuple is the ordered collection of element in python


It is immutable means cannot be changed after creation

Tuple is repesented by ( )



```python
T=(10,20,30)
print(T)
```

Set:-

Set is unordered collection of element

It is mutable means can be changed

Duplicate value are not allowed(does not print duplicate value)

Set is represented by { }

```python
S={10,20,30,20}
print(S)
```

Dictionary:-

Dictionary is the collection of the data stored in key-value pair

It is mutable means can be modified

Duplicate value are allowed

Dictionary is represented by { }

```python
D={"name":"Rajashri","age":21}
print(D)
```

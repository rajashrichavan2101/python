## There are two types of type conversion in the data type

1.Implicit:-

## It automatically convert one data type into the another data type

2.Explicit:-

## In explicit it is the process of manually converting one data type into the another data type

It is also called type casting


```python
L=[10,20,30]
T=tuple(L)
print(T)
```

```python
x={10,20,30}
y=tuple(x)
print(y)
```

```python
A=[10,20,30]
B=set(A)
print(B)
```

```python
T=(10,20,30)
S=set(T)
print(S)
```
